```python
import pandas as pd

def one_hot_encode(df, column_name):
  """
  Performs one-hot encoding on a specified column of a Pandas DataFrame.

  Args:
    df (pd.DataFrame): The input DataFrame.
    column_name (str): The name of the column to be one-hot encoded.

  Returns:
    pd.DataFrame: A new DataFrame with the original column dropped and
                  new binary columns for each unique value in the original
                  column.
  """
  encoded_df = pd.get_dummies(df, columns=['company','fuel_type'], prefix=column_name)
  return encoded_df

# Example usage:
data = {'Category': ['A', 'B', 'A', 'C', 'B'],
        'Value': [10, 20, 15, 25, 30]}
df = pd.DataFrame(car)
print("Original DataFrame:")
print(df)

car= one_hot_encode(car, 'Category')
print("\nOne-Hot Encoded DataFrame:")
print(car)
```
